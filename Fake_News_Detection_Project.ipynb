```python
import pandas as pd
import numpy as np
```

```python
true = pd.read_csv('/content/drive/MyDrive/True.csv')
fake = pd.read_csv('/content/drive/MyDrive/Fake.csv')
```

```python
true
```

```python
# Labelling the real news as 1 and fake as 0
true['Label'] = 1
fake['Label'] = 0
```

```python
# Combining the first 5000 data from both datasets
comb = [true.loc[:5000][:] , fake.loc[:5000][:]] 
df = pd.concat(comb)
```

```python
X = df.drop('Label', axis = 1)
y = df['Label']
```

```python
# Delete missing data
df.dropna()
```

```python
# Make a copy of df 
df2 = df.copy()
df2.reset_index(inplace = True)
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
import nltk
```

```python
ps = PorterStemmer()
nltk.download('stopwords')
corpus = []

for i in range(len(df2)):
  review = re.sub('[^a-zA-Z]', ' ', df2['text'][i]) #remove punctuations
  # review = df2['text'][i]
  review = review.lower()
  review = review.split()

#removing stopwords
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid_v = TfidfVectorizer(max_features = 5000, ngram_range = (1,3))
X = tfid_v.fit_transform(corpus).toarray()
y = df2["Label"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=16)
logreg.fit(X_train, y_train)
pred = logreg.predict(X_test)
```

```python
from sklearn import metrics
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```
